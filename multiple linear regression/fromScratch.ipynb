## IMPORTING DATASET

```python
import pandas as pd
import numpy as np
```

## LOADING DATASET

```python
dataset=pd.read_csv('Train.csv')
```

```python
dataset.head()
```

## GETTING X AND y VALUES FROM THE DATASET

```python
X=dataset.iloc[:,:-1].values
ones=np.ones([X.shape[0],1])      #vector with number of rows equal to number of rows equal to X 
X=np.concatenate((ones,X),axis=1) #This concatenation helps us deal with the constant term in the parameter matrix and 
                                  #our gradient descent is able to perform its action using a single equation
```

```python
print(X)
print(ones)
```

```python
y=dataset.iloc[:,-1:].values
print(y)
```

## SETTING HYPERPARAMETERS

```python
theta=np.zeros([1,X.shape[1]])      #randomly initiliazed the values of coefficients to 0
#print(theta)             
alpha=0.01                #learning rate
iters=1000                #number of times gradient descent works
```

## COST FUNCTION

```python
def computeCost(X,y,theta):
    toBeSummed=np.power(((X@theta.T)-y),2)
    return np.sum(toBeSummed)/(2*len(X))
```

## CURRENT VALUE OF COST FUNCTION

```python
print(computeCost(X,y,theta))    #we have to minimize this value using gradient descent
```

## GRADIENT DESCENT FUNCTION

```python
def gradientDescent(X,y,theta,iters,alpha):
    cost=np.zeros(iters)
    for i in range(iters):
        theta=theta-(alpha/len(X))*np.sum((X*(X@theta.T-y)),axis=0)     
        cost[i]=computeCost(X,y,theta)
    return theta,cost
```

## MODEL IN ACTION

```python
g,cost=gradientDescent(X,y,theta,iters,alpha)
print(g)
```

```python
finalCost=computeCost(X,y,g)
print(finalCost)
```
