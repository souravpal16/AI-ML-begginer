```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv('Housing_Prices.csv')
data.rename(columns={'Y house price of unit area':'price','X4 number of convenience stores':'conv.stores','X3 distance to the nearest MRT station':'dist.MRT'},inplace=True)
data.head()
```

```python
data.describe()
```

<p>1.At a first glance "transaction date" doesn't seem to be a feature which we should consider but observation no. 3,4 suggest otherwise which makes it a necessary to involve in our model.<br>
2.House age, distance to the nearest MRT station, no. of convenience are definitely the factors responsible for housing prices.<br>
3.The maximum and minimum values of latitde and longitude suggest that these houses being considered are in the same city but definitely not the same locality otherwise no. of convience stores would have been nearly same for all houses. How exactly latitude and longitude affect the housing prices would be more clear from the visualisation. </p>

<h1>DATA VISUALISATION</h1>

```python
import seaborn as sns
```

```python
plt.figure(figsize=(10,10))
sns.scatterplot(x=data['X1 transaction date'],y=data['price'])
```

```python
plt.figure(figsize=(10,10))
sns.scatterplot(x=data['X2 house age'],y=data['price'])
```

```python
plt.figure(figsize=(10,10))
sns.scatterplot(x=data['dist.MRT'],y=data['price'])
```

```python
plt.figure(figsize=(10,10))
sns.scatterplot(x=data['conv.stores'],y=data['price'])
```

```python
plt.figure(figsize=(10,10))
sns.scatterplot(x=data['X5 latitude'],y=data['price'])
```

```python
plt.figure(figsize=(10,10))
sns.scatterplot(x=data['X6 longitude'],y=data['price'])
```

```python
plt.figure(figsize=(10,10))
sns.jointplot(data['X2 house age'],data['dist.MRT'],kind='kde')
```

```python
plt.figure(figsize=(10,10))
sns.jointplot(data['X5 latitude'],data['X6 longitude'],kind='kde')
```

```python
plt.figure(figsize=(10,10))
sns.jointplot(data['conv.stores'],data['dist.MRT'],kind='kde')
```

<p>1.None of the graphs raise any hope for simple linear regression.<br>
2.No. of convenience stores and transaction dates play little to no role in predicting the prices. So we are going
to make models on multiple linear regression, support vector regression, random forest trees each of which would be 
trained with and without said features.<br>
3.The joint plots suggest that major of the houses are close to each other mainly central city and a few houses are loacted on the outskirts of the city which explain why some of the houses have high distance from the MRT station and simultaenoulsy few convenience stores.</p>
